```python
train_data=genfromtxt("/content/My_Drive/MyDrive/Data/PRNN/Assignment 1/p3_train.csv",delimiter=",")
test_data=genfromtxt("/content/My_Drive/MyDrive/Data/PRNN/Assignment 1/p3_test.csv",delimiter=",")

x_train,y_train=train_data[0:60000,0:10],train_data[0:60000,10]
x_test,y_test=test_data[:,0:10],test_data[:,10]

#y_train=reshape(y_train,[len(x_train[:,0]),1])
#y_train.shape

N,d=x_train.shape
S,d=x_test.shape

labels=[1,2,3,4,5]
k=len(labels)
print("N=",N,"d=",d)
#with Normal Class Conditional Densities:
#y_train
```

```python
#Required Order Statistics!!!

temp=[]

for i in labels:
  for j in range(N):
    if y_train[j]==i:
      temp.append(i)

c1,c2,c3,c4,c5=0,0,0,0,0

for i in temp:
  if i==1:
   c1=c1+1
  elif i==2:
    c2=c2+1
  elif i==3:
    c3=c3+1
  elif i==4:
    c4=c4+1
  elif i==5:
    c5=c5+1

#Class Priors

Py1,Py2,Py3,Py4,Py5=c1/N,c2/N,c3/N,c4/N,c5/N

print("Py1=",Py1,"Py2=",Py2,"Py3=",Py3,"Py4=",Py4,"Py5=",Py5)

Py=[Py1,Py2,Py3,Py4,Py5]

#sample means:
sample_means=[]
sum1,sum2,sum3,sum4,sum5=0,0,0,0,0

for i in range(N):

  if y_train[i]==1:
    sum1=sum1+x_train[i,:]
  elif y_train[i]==2:
    sum2=sum2+x_train[i,:]
  elif y_train[i]==3:
    sum3=sum3+x_train[i,:]
  elif y_train[i]==4:
    sum4=sum4+x_train[i,:]
  elif y_train[i]==5:
    sum5=sum5+x_train[i,:]

sample_mean1,sample_mean2,sample_mean3,sample_mean4,sample_mean5=sum1/c1,sum2/c2,sum3/c3,sum4/c4,sum5/c5
sample_mean1,sample_mean2,sample_mean3,sample_mean4,sample_mean5=reshape(sample_mean1,[d,1]),reshape(sample_mean2,[d,1]),reshape(sample_mean3,[d,1]),reshape(sample_mean4,[d,1]),reshape(sample_mean5,[d,1])
print("sample_mean1=",sample_mean1,"sample_mean2=",sample_mean2,"sample_mean3=",sample_mean3,
      "sample_mean4=",sample_mean4,"sample_mean5=",sample_mean5)

#sample Covariances:

#for i in labels:

#X_1,X_2,X_3,X_4,X_5=[],[],[],[],[]
sum_cov1,sum_cov2,sum_cov3,sum_cov4,sum_cov5=0,0,0,0,0
for i in range(N):
  if y_train[i]==1:
    x_train_temp=reshape(x_train[i,:],[d,1])
    sum_cov1=sum_cov1+(x_train_temp-sample_mean1) @ (x_train_temp-sample_mean1).T
  elif y_train[i]==2:
    x_train_temp=reshape(x_train[i,:],[d,1])
    sum_cov2=sum_cov2+(x_train_temp-sample_mean2) @ (x_train_temp-sample_mean2).T
  elif y_train[i]==3:
    x_train_temp=reshape(x_train[i,:],[d,1])
    sum_cov3=sum_cov3+(x_train_temp-sample_mean3) @ (x_train_temp-sample_mean3).T
  elif y_train[i]==4:
    x_train_temp=reshape(x_train[i,:],[d,1])
    sum_cov4=sum_cov4+(x_train_temp-sample_mean4) @ (x_train_temp-sample_mean4).T
  elif y_train[i]==5:
    x_train_temp=reshape(x_train[i,:],[d,1])
    sum_cov5=sum_cov5+(x_train_temp-sample_mean5) @ (x_train_temp-sample_mean5).T

#X_1,X_2,X_3,X_4,X_5=array(X_1),array(X_2),array(X_3),array(X_4),array(X_5)

cov1,cov2,cov3,cov4,cov5=sum_cov1/c1,sum_cov2/c2,sum_cov3/c3,sum_cov4/c4,sum_cov5/c5

print("cov1=",cov1,"cov2=",cov2,"cov3=",cov3,"cov4=",cov4,"cov5=",cov5)

Cov=[cov1,cov2,cov3,cov4,cov5]
Sample_Mean=[sample_mean1,sample_mean2,sample_mean3,sample_mean4,sample_mean5]
inv_cov1,inv_cov2,inv_cov3,inv_cov4,inv_cov5=linalg.inv(cov1),linalg.inv(cov2),linalg.inv(cov3),linalg.inv(cov4),linalg.inv(cov5)

Inv_Cov=[inv_cov1,inv_cov2,inv_cov3,inv_cov4,inv_cov5]

print("inv_cov1=",inv_cov1,"inv_cov2=",inv_cov2,"inv_cov3=",inv_cov3,"inv_cov4=",inv_cov4,"inv_cov5=",inv_cov5)
```

```python
#Class Conditional Densities--->Normal Densities

def multivariate_Gaussian_distribution(x,class_label,sample_mean,covariance,covariance_inverse):

  #print(shape(x))
  d=shape(x)[0]
  
  
  constant=((2*pi)**d/2)*(linalg.det(covariance)**(1/2))
  #covariance_inverse=linalg.inv(covariance)
  #if class_label==1:
  #print((x-sample_mean).shape,covariance_inverse.shape)
  
  #f_xgiveny=(1/constant)*exp((-dot((x-sample_mean),matmul(covariance_inverse,(x-sample_mean))))/2)
  f_xgiveny=(1/constant)*exp(-(1/2)*(x-sample_mean).T@covariance_inverse@(x-sample_mean))
  # #if class_label==2

  return f_xgiveny

fxy_mat_train=zeros((N,len(labels)))

for i in (labels):
  for j in range(N):
    x_train_temp=reshape(x_train[j,:],[d,1])
    fxy_mat_train[j,i-1]=multivariate_Gaussian_distribution(x_train_temp,i,Sample_Mean[i-1],Cov[i-1],Inv_Cov[i-1])

fyx_mat_train=zeros((N,len(labels)))

for i in (labels):
  for j in range(N):
    fyx_mat_train[j,i-1]=fxy_mat_train[j,i-1]*Py[i-1]

print("fxy_mat_train",fxy_mat_train)

#Y_predictions:

Y_pred=[]

for j in range(N):
  #for i in labels:
  max_temp=max(fyx_mat_train[j,:])
  index_m=where(fyx_mat_train[j,:]==max_temp)
  Y_pred.append(array(index_m)+1)

Y_pred=array(Y_pred)

Y_pred
```

```python
c=0
for i in range(N):
  if Y_pred[i]-y_train[i]==0:
    c=c+1

accuracy_train=(c/N)*100
print("accuracy_train=",accuracy_train)
```
